```python
import requests
import pandas as pd

url = "https://berita-indo-api.vercel.app/v1/cnn-news"
response = requests.get(url)

data = response.json()
news_data = data['data']

df = pd.DataFrame(news_data)
df
```

```python
df = df[['title', 'link', 'contentSnippet', 'isoDate']]

df = df.rename(columns={
    'title': 'news_title',
    'link': 'news_link',
    'contentSnippet': 'news_summary',
    'isoDate': 'published_at'
})
```

```python
df = df.drop_duplicates(subset='news_link')
```

```python
df = df.dropna(subset=['news_title', 'news_summary'])
```

```python
df['news_title'] = df['news_title'].str.title()
df['news_summary'] = df['news_summary'].str.lower()
```

```python
df['title_length'] = df['news_title'].str.len()
df['word_count'] = df['news_summary'].str.split().str.len()
```

```python
df = df[df['word_count'] > 5]
```

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    user="postgres",
    password="kukang12",
    dbname="assigment",
    port=5432
)

cursor = conn.cursor()

cursor.execute("TRUNCATE TABLE stg_news")

insert_query = """
INSERT INTO stg_news
(news_link, news_title, news_summary, published_at, title_length, word_count)
VALUES (%s, %s, %s, %s, %s, %s)
"""

for _, row in df.iterrows():
    cursor.execute(insert_query, (
        row['news_link'],
        row['news_title'],
        row['news_summary'],
        row['published_at'],
        row['title_length'],
        row['word_count']
    ))

conn.commit()
cursor.close()
conn.close()
```
